```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig=px.bar(x=["a", "a", "b", 3], y = [1, 2, 3, 4])
fig.update_xaxes(type='category')

fig.show()
```

```python
df=px.data.tips()
fig=px.bar(df, x="day", y="total_bill",
           color="smoker", barmode="group", facet_col="sex",
           category_orders={"day": ["Thur", "Fri", "Sat", "Sun"],
                            "smoker": ["Yes", "No"],
                            "sex": ["Male", "Female"]},
           labels={'total_bill': 'total bill'}
)

fig.show()
```

```python
x=['b', 'a', 'c', 'd']

fig=go.Figure()

fig.add_trace(go.Bar(x=x, y=[2, 5, 1, 9], name='Montreal'))
fig.add_trace(go.Bar(x=x, y=[1, 4, 9, 16], name='Ottawa'))
fig.add_trace(go.Bar(x=x, y=[6, 8, 4.5, 8], name='Toronto'))

fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='category ascending')

fig.show()
```

```python

x=['b', 'a', 'c', 'd']

fig=go.Figure()

fig.add_trace(go.Bar(x=x, y=[2, 5, 1, 9], name='Montreal'))
fig.add_trace(go.Bar(x=x, y=[1, 4, 9, 16], name='Ottawa'))
fig.add_trace(go.Bar(x=x, y=[6, 8, 4.5, 8], name='Toronto'))

fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='total descending')

fig.show()
```

```python

x=['b', 'a', 'c', 'd']

fig=go.Figure()

fig.add_trace(go.Bar(x=x, y=[2, 5, 1, 9], name='Montreal'))
fig.add_trace(go.Bar(x=x, y=[1, 4, 9, 16], name='Ottawa'))
fig.add_trace(go.Bar(x=x, y=[6, 8, 4.5, 8], name='Toronto'))

fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='array',
                 categoryarray=['b', 'a', 'd', 'c'])

fig.show()
```

```python
fig=px.bar(x=["A", "B", "C"], y=[1, 3, 2])

fig.show()
```

```python
fig=px.bar(x=["A", "B", "C"], y=[1, 3, 2])
fig.update_xaxes(showgrid=True, ticks="outside")

fig.show()
```

```python
fig=px.bar(x=["A", "B", "C"], y=[1, 3, 2])
fig.update_xaxes(showgrid=True, ticks="outside", tickson="boundaries")

fig.show()
```

```python
fig=go.Figure()

fig.add_trace(go.Bar(
    x=[['First', 'First', 'Second', 'Second'],
       ["A", "B", "A", "B"]],
    y=[2, 3, 1, 5],
    name="Adults",
))

fig.add_trace(go.Bar(
    x=[['First', 'First', 'Second', 'Second'],
       ["A", "B", "A", "B"]],
    y=[8, 3, 6, 5],
    name="Children",
))

fig.update_layout(title_text="Multi-category axis")

fig.show()
```
